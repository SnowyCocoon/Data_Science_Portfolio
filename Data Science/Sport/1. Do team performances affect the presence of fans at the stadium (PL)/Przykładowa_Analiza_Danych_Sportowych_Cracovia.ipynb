```python
import pandas as pd
import seaborn as sns
```

```python
CRA_df_14_15 = pd.read_csv("spotkaniafw20142015k3.txt", sep='\t', names=['Data','Godzina','Gospodarz','Gość','Wynik','Miejscowość','Temperatura','Wilgotność','Prędkość wiatru','Opad','Mgła','Burza','Grają u siebie','Frekwencja','Wygrana'
])

CRA_df_15_16 = pd.read_csv("spotkaniafw20152016k3.txt", sep='\t', names=['Data','Godzina','Gospodarz','Gość','Wynik','Miejscowość','Temperatura','Wilgotność','Prędkość wiatru','Opad','Mgła','Burza','Grają u siebie','Frekwencja','Wygrana'
])

CRA_df_16_17 = pd.read_csv("spotkaniafw20162017k3.txt", sep='\t', names=['Data','Godzina','Gospodarz','Gość','Wynik','Miejscowość','Temperatura','Wilgotność','Prędkość wiatru','Opad','Mgła','Burza','Grają u siebie','Frekwencja','Wygrana'
])

CRA_df_17_18 = pd.read_csv("spotkaniafw20172018k3.txt", sep='\t', names=['Data','Godzina','Gospodarz','Gość','Wynik','Miejscowość','Temperatura','Wilgotność','Prędkość wiatru','Opad','Mgła','Burza','Grają u siebie','Frekwencja','Wygrana'
])
```

```python
wyniki_frekwencja_14_15 = CRA_df_14_15[['Data', 'Godzina', 'Gospodarz', 'Gość', 'Wynik', 'Miejscowość', 'Grają u siebie', 'Frekwencja', 'Wygrana', 'Temperatura', 'Opad']]
wyniki_frekwencja_15_16 = CRA_df_15_16[['Data', 'Godzina', 'Gospodarz', 'Gość', 'Wynik', 'Miejscowość', 'Grają u siebie', 'Frekwencja', 'Wygrana', 'Temperatura', 'Opad']]
wyniki_frekwencja_16_17 = CRA_df_16_17[['Data', 'Godzina', 'Gospodarz', 'Gość', 'Wynik', 'Miejscowość', 'Grają u siebie', 'Frekwencja', 'Wygrana', 'Temperatura', 'Opad']]
wyniki_frekwencja_17_18 = CRA_df_17_18[['Data', 'Godzina', 'Gospodarz', 'Gość', 'Wynik', 'Miejscowość', 'Grają u siebie', 'Frekwencja', 'Wygrana', 'Temperatura', 'Opad']]
```

```python
wyniki_frekwencja_14_15['Czy Wygrał Poprzedni'] = 0
wyniki_frekwencja_15_16['Czy Wygrał Poprzedni'] = 0
wyniki_frekwencja_16_17['Czy Wygrał Poprzedni'] = 0
wyniki_frekwencja_17_18['Czy Wygrał Poprzedni'] = 0

wyniki_frekwencja_14_15['Czy Wygrał Dwa Poprzednie'] = 0
wyniki_frekwencja_15_16['Czy Wygrał Dwa Poprzednie'] = 0
wyniki_frekwencja_16_17['Czy Wygrał Dwa Poprzednie'] = 0
wyniki_frekwencja_17_18['Czy Wygrał Dwa Poprzednie'] = 0

wyniki_frekwencja_14_15['Czy Wygrał Trzy Poprzednie'] = 0
wyniki_frekwencja_15_16['Czy Wygrał Trzy Poprzednie'] = 0
wyniki_frekwencja_16_17['Czy Wygrał Trzy Poprzednie'] = 0
wyniki_frekwencja_17_18['Czy Wygrał Trzy Poprzednie'] = 0

wyniki_frekwencja_14_15['Stopień Opadu'] = 0
wyniki_frekwencja_15_16['Stopień Opadu'] = 0
wyniki_frekwencja_16_17['Stopień Opadu'] = 0
wyniki_frekwencja_17_18['Stopień Opadu'] = 0

wyniki_frekwencja_14_15['Poziom Temperatur'] = 0
wyniki_frekwencja_15_16['Poziom Temperatur'] = 0
wyniki_frekwencja_16_17['Poziom Temperatur'] = 0
wyniki_frekwencja_17_18['Poziom Temperatur'] = 0
```

```python
def complete1(i, df):
  if(df['Wygrana'][i+1] == 1):
    df['Czy Wygrał Poprzedni'][i] = "Wygrana"
  elif(df['Wygrana'][i+1] == 2):
    df['Czy Wygrał Poprzedni'][i] = "Remis"
  else:
    df['Czy Wygrał Poprzedni'][i] = "Przegrana"

for i in range (0,29):
  complete1(i, wyniki_frekwencja_14_15)
  complete1(i, wyniki_frekwencja_15_16)
  complete1(i, wyniki_frekwencja_16_17)
  complete1(i, wyniki_frekwencja_17_18)
```

```python
def complete2(i, df):
  if(df['Wygrana'][i+1] == 1 and df['Wygrana'][i+2] == 1):
    df['Czy Wygrał Dwa Poprzednie'][i] = "Tak"
  else:
    df['Czy Wygrał Dwa Poprzednie'][i] = "Nie"

for i in range (0,28):
  complete2(i, wyniki_frekwencja_14_15)
  complete2(i, wyniki_frekwencja_15_16)
  complete2(i, wyniki_frekwencja_16_17)
  complete2(i, wyniki_frekwencja_17_18)
```

```python
def complete3(i, df):
  if(df['Wygrana'][i+1] == 1 and df['Wygrana'][i+2] == 1 and df['Wygrana'][i+3] == 1):
    df['Czy Wygrał Trzy Poprzednie'][i] = "Tak"
  else:
    df['Czy Wygrał Trzy Poprzednie'][i] = "Nie"

for i in range (0,27):
  complete3(i, wyniki_frekwencja_14_15)
  complete3(i, wyniki_frekwencja_15_16)
  complete3(i, wyniki_frekwencja_16_17)
  complete3(i, wyniki_frekwencja_17_18)
```

```python
def complete4(i, df):
  if(df['Opad'][i] == 0):
    df['Stopień Opadu'][i] = "Brak"
  elif(df['Opad'][i] <= 2.5):
    df['Stopień Opadu'][i] = "Lekki Opad"
  elif(df['Opad'][i] <= 7.5):
    df['Stopień Opadu'][i] = "Umiarkowany Opad"
  else:
    df['Stopień Opadu'][i] = "Silny Opad"

for i in range (0,30):
  complete4(i, wyniki_frekwencja_14_15)
  complete4(i, wyniki_frekwencja_15_16)
  complete4(i, wyniki_frekwencja_16_17)
  complete4(i, wyniki_frekwencja_17_18)
```

```python
def complete5(i, df):
  if(df['Temperatura'][i] <= 0):
    df['Poziom Temperatur'][i] = "Bardzo zimno"
  elif(df['Temperatura'][i] <= 10):
    df['Poziom Temperatur'][i] = "Zimno"
  elif(df['Temperatura'][i] <= 25):
    df['Poziom Temperatur'][i] = "Umiarkowanie/Ciepło"
  else:
    df['Poziom Temperatur'][i] = "Gorąco"

for i in range (0,30):
  complete5(i, wyniki_frekwencja_14_15)
  complete5(i, wyniki_frekwencja_15_16)
  complete5(i, wyniki_frekwencja_16_17)
  complete5(i, wyniki_frekwencja_17_18)
```

```python
final_df = pd.concat([wyniki_frekwencja_14_15, wyniki_frekwencja_15_16, wyniki_frekwencja_16_17, wyniki_frekwencja_17_18], ignore_index=True)
```

```python
final_df = final_df.loc[final_df['Gospodarz'] == 'CRA']
```

```python
final_df = final_df.drop([84], axis=0) #Brak frekwencji
final_df = final_df.drop([20], axis=0) #Odstający wynik
final_df = final_df.drop([33], axis=0) #Odstający wynik
final_df = final_df.drop([58], axis=0) #Odstający wynik
final_df = final_df.drop([60], axis=0) #Odstający wynik
final_df = final_df.drop([86], axis=0) #Odstający wynik
final_df = final_df.drop([100], axis=0) #Odstający wynik
final_df = final_df.drop([98], axis=0) #Odstający wynik
final_df
```

```python
final_df["Frekwencja"] =  [int(x) for x in final_df["Frekwencja"] if 1]
```
